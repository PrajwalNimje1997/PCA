```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("/Users/prajwalite/Downloads/heart.csv")
df.head()
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaledData = scaler.transform(df)
```

```python
scaledData
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca.fit(scaledData)
```

```python
pcaTransformed = pca.transform(scaledData)
```
